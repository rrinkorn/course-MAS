````python
"""
Подсчёт использованных токенов.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 10: Подсчёт токенов")
print("=" * 60)

from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0,
)

print("\n10.1. Токены после вызова (usage_metadata):")
result = model.invoke("Напиши короткое стихотворение о программировании")

if hasattr(result, "usage_metadata") and result.usage_metadata:
    print(f"  Input tokens: {result.usage_metadata.get('input_tokens', 'N/A')}")
    print(f"  Output tokens: {result.usage_metadata.get('output_tokens', 'N/A')}")
    print(f"  Total tokens: {result.usage_metadata.get('total_tokens', 'N/A')}")
else:
    print("  usage_metadata недоступен")

print("\n10.2. Предварительный подсчёт (get_num_tokens):")
text = "Это текст для подсчёта токенов на русском языке"
try:
    num_tokens = model.get_num_tokens(text)
    print(f"  Текст: '{text}'")
    print(f"  Токенов: {num_tokens}")
except Exception as e:
    print(f"  Ошибка: {e}")
````

````python
"""
Callback для отслеживания использования.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 11: Callback для отслеживания токенов")
print("=" * 60)

from langchain_openai import ChatOpenAI
from langchain_core.callbacks import BaseCallbackHandler


class TokenTracker(BaseCallbackHandler):
    """Отслеживает использование токенов."""

    def __init__(self):
        self.total_input = 0
        self.total_output = 0
        self.calls = 0

    def on_llm_end(self, response, **kwargs):
        self.calls += 1
        if hasattr(response, "llm_output") and response.llm_output:
            usage = response.llm_output.get("token_usage", {})
            self.total_input += usage.get("prompt_tokens", 0)
            self.total_output += usage.get("completion_tokens", 0)

    def report(self):
        return {
            "calls": self.calls,
            "total_input_tokens": self.total_input,
            "total_output_tokens": self.total_output,
            "total_tokens": self.total_input + self.total_output,
        }


tracker = TokenTracker()
model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0,
    callbacks=[tracker],
)

# Несколько вызовов
model.invoke("Привет!")
model.invoke("Как дела?")
model.invoke("Что такое Python?")

print("\n11.1. Отчёт об использовании:")
report = tracker.report()
for key, value in report.items():
    print(f"  {key}: {value}")
````

````python
"""
Оценка стоимости вызовов.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 12: Оценка стоимости")
print("=" * 60)

# Примерные цены (могут меняться!)
PRICES = {
    "gpt-4o": {"input": 2.50, "output": 10.00},  # за 1M токенов
    "gpt-4o-mini": {"input": 0.15, "output": 0.60},
    "gpt-4-turbo": {"input": 10.00, "output": 30.00},
    "claude-3-5-sonnet": {"input": 3.00, "output": 15.00},
}


def estimate_cost(model_name: str, input_tokens: int, output_tokens: int) -> float:
    """Оценивает стоимость в долларах."""
    if model_name not in PRICES:
        return 0.0
    prices = PRICES[model_name]
    input_cost = (input_tokens / 1_000_000) * prices["input"]
    output_cost = (output_tokens / 1_000_000) * prices["output"]
    return input_cost + output_cost


print("\n12.1. Сравнение стоимости моделей (1000 input + 500 output токенов):")
for model_name in PRICES:
    cost = estimate_cost(model_name, 1000, 500)
    print(f"  {model_name}: ${cost:.6f}")

print("\n12.2. Стоимость при масштабе (1M запросов по 1000+500 токенов):")
for model_name in PRICES:
    cost = estimate_cost(model_name, 1_000_000_000, 500_000_000)
    print(f"  {model_name}: ${cost:,.2f}")
````

````python
"""
Базовый стриминг с методом stream().
"""
print("\n" + "=" * 60)
print("ПРИМЕР 16: Базовый стриминг")
print("=" * 60)

from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0.7,
)

print("\n16.1. Потоковый вывод:")
print("  ", end="")
for chunk in model.stream("Напиши короткое хайку о программировании"):
    print(chunk.content, end="", flush=True)
print()
````

````python
"""
Асинхронный стриминг.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 17: Асинхронный стриминг")
print("=" * 60)

from langchain_openai import ChatOpenAI


async def stream_response():
    model = ChatOpenAI(
        base_url=os.getenv("OPENROUTER_BASE_URL"),
        api_key=os.getenv("OPENROUTER_API_KEY"),
        model="gpt-4o-mini",
        temperature=0.7,
    )

    print("\n17.1. Async стриминг:")
    print("  ", end="")
    async for chunk in model.astream("Расскажи короткую шутку о Python"):
        print(chunk.content, end="", flush=True)
    print()


asyncio.run(stream_response())
````

````python
"""
Стриминг через цепочку LCEL.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 18: Стриминг через цепочку")
print("=" * 60)

from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0.7,
)
prompt = ChatPromptTemplate.from_template("Напиши лимерик о {topic}")
chain = prompt | model | StrOutputParser()

print("\n18.1. Стриминг из цепочки:")
print("  ", end="")
for chunk in chain.stream({"topic": "программистах"}):
    print(chunk, end="", flush=True)
print()
````

````python
"""
Метод bind() для привязки параметров.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 19: Model.bind()")
print("=" * 60)

from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0,
)

# Привязываем параметры
json_model = model.bind(response_format={"type": "json_object"})

prompt = ChatPromptTemplate.from_template("Верни информацию о {animal} в формате JSON с полями: name, type, habitat")

chain = prompt | json_model | StrOutputParser()

print("\n19.1. Модель с привязанным JSON режимом:")
result = chain.invoke({"animal": "пингвин"})
print(f"  {result}")

# Модель с stop-последовательностью
code_model = model.bind(stop=["```"])
print("\n19.2. Модель с привязанной stop-последовательностью:")
result = code_model.invoke("Напиши код: print hello world в Python, в блоке ```python")
print(f"  {result.content}")
````

````python
"""
ConfigurableField для runtime-параметризации.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 20: Configurable Model")
print("=" * 60)

from langchain_openai import ChatOpenAI
from langchain_core.runnables import ConfigurableField

# Модель с настраиваемой temperature
model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0.5,
).configurable_fields(
    temperature=ConfigurableField(
        id="temp",
        name="Temperature",
        description="Контроль креативности",
    )
)

prompt = ChatPromptTemplate.from_template("Придумай слоган для {product}")
chain = prompt | model | StrOutputParser()

print("\n20.1. Вызов с temperature=0.1:")
result1 = chain.invoke({"product": "кофейня"}, config={"configurable": {"temp": 0.1}})
print(f"  {result1}")

print("\n20.2. Вызов с temperature=0.9:")
result2 = chain.invoke({"product": "кофейня"}, config={"configurable": {"temp": 0.9}})
print(f"  {result2}")
````

````python
"""
ConfigurableAlternatives для переключения моделей.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 21: Configurable Alternatives")
print("=" * 60)

from langchain_openai import ChatOpenAI
from langchain_core.runnables import ConfigurableField

# Модель с альтернативами
model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0,
).configurable_alternatives(
    ConfigurableField(id="model_type"),
    default_key="balanced",
    creative=ChatOpenAI(
        base_url=os.getenv("OPENROUTER_BASE_URL"),
        api_key=os.getenv("OPENROUTER_API_KEY"),
        model="gpt-4o-mini",
        temperature=0.9,
    ),
    precise=ChatOpenAI(
        base_url=os.getenv("OPENROUTER_BASE_URL"),
        api_key=os.getenv("OPENROUTER_API_KEY"),
        model="gpt-4o-mini",
        temperature=0,
    ),
)

prompt = ChatPromptTemplate.from_template("Опиши {object} одним предложением")
chain = prompt | model | StrOutputParser()

print("\n21.1. Режим 'balanced' (по умолчанию):")
result = chain.invoke({"object": "облако"})
print(f"  {result}")

print("\n21.2. Режим 'creative':")
result = chain.invoke({"object": "облако"}, config={"configurable": {"model_type": "creative"}})
print(f"  {result}")

print("\n21.3. Режим 'precise':")
result = chain.invoke({"object": "облако"}, config={"configurable": {"model_type": "precise"}})
print(f"  {result}")
````

````python
"""
Управление rate limiting при батчинге.
"""
print("\n" + "=" * 60)
print("ПРИМЕР 23: Rate limiting при батчинге")
print("=" * 60)

from langchain_openai import ChatOpenAI
import time

model = ChatOpenAI(
    base_url=os.getenv("OPENROUTER_BASE_URL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    model="gpt-4o-mini",
    temperature=0,
)
prompt = ChatPromptTemplate.from_template("Переведи на английский: {text}")
chain = prompt | model | StrOutputParser()

texts = [{"text": word} for word in ["привет", "мир", "код", "данные", "модель"]]

print("\n23.1. Batch с max_concurrency=2:")
start = time.time()
results = chain.batch(
    inputs=texts,
    config={"max_concurrency": 2},
)
elapsed = time.time() - start

for inp, out in zip(texts, results):
    print(f"  '{inp['text']}' -> '{out}'")
print(f"\n  Время: {elapsed:.2f}s")

print("\n  СОВЕТ: Ограничивайте concurrency для соблюдения rate limits провайдера")
````
